```python
import pandas as pd
import numpy as np

# Load the provided CSV dataset
file_path = "dataset/istanbul_satilik_evler_2023.csv"
df = pd.read_csv(file_path)

# Display the first few rows of the dataframe to understand its structure
df.head()
```

```python
# Clean and preprocess the data

# Strip any leading/trailing spaces from column names
df.columns = df.columns.str.strip()

# Convert 'Price' to numeric
df['Price'] = pd.to_numeric(df['Price'].str.replace(",", "").str.strip(), errors='coerce')

# Convert 'Room' to numeric (extract the number of rooms)
df['Room'] = df['Room'].str.split('+').str[0].str.strip()
df['Room'] = pd.to_numeric(df['Room'], errors='coerce')

# Convert other columns to appropriate data types
df['Area'] = pd.to_numeric(df['Area'], errors='coerce')
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')
df['Floor'] = pd.to_numeric(df['Floor'], errors='coerce')

# Drop rows with missing values
df.dropna(inplace=True)

# Display the cleaned dataframe
df.head()
```
